# Function

```python
# functions
# keyword def.
# the name of function followed by a set of paranthesis and a colon.
# a docstring in triple quotes describes the function.
# to call a function, give the name of the function followed by set of paranthesis.
def greet_user():
    """display a simple greeting."""
    print("hello.")
greet_user()

# defined function
def compute(x: int, y: float) -> None:
  x + y
def compute2(x: int, y: int) -> int:
  return x + y

# Recursion
def dec_to_bin(n):
  if n < 2:
    print(n, end = '')
    return
  dec_to_bin(n // 2)
  print(n % 2, end = '')

dec_to_bin(10)
```

```python
# passing information:
# information passed to a function are called arguments, received are called parameters.
# arguments are icluded in the paranthesis after the function's name .
# parameter's are listed in the functions definition.
def greet_user(username):
    """Display a simple greeting."""
    print("hello,"+ username + "!")
greet_user('jessie')
```

```python
# positional and keyword arguments
def describe_pet(animal, name):
    """Display information about a pet."""
    print("\nI have a "+ animal + ".")
    print("its name is " + name + ".")
# using positional arg:
describe_pet('hamster', 'harry')

# using keyword arguments:
describe_pet(name = 'willie', animal= 'dog')
```

```python
# default values
# when function call ignores argument , default values will be used.
def describe_pet2(name, animal='dog'):
    """Display information about a pet."""
    print("\nI have a "+ animal + ".")
    print("its name is " + name + ".")

describe_pet2('harry', 'hamster')
describe_pet2('willie')
```

```python
# return values
# returning a single value:
def get_full_name(first, last):
    """Return the neatly formated full name."""
    full_name = first + ' '+ last
    return full_name.title()

musician = get_full_name('jimi', 'hendrix')
print(musician)


# return a dictionary
def build_person(first, last):
    """Return a dictionary of information about a person."""
    person = {'first' : first, 'last': last}
    return person

musician = build_person('jimi', 'hendrix')
print(musician)


# return a dictionary with optional values:
def build_person(first, last, age = None):
    """return a dictionary of information about a person. """
    person = {'first': first, 'last': last}
    if age:
        person['age'] = age
    return person

musician = build_person('jimi', 'hendrix', 27)
print(musician)

musician= build_person('janis', 'joplin')
print(musician)
```

```python
# Pass a list to a function:
# can pass a list as an argument to a function, and the function can work with the values in the list.
# any changes the function makes to a list will affect the original list.
# you can prevent by passing a copy of list as an argument.
def greet_users(names):
    for name in names:
        msg = "Hello, " + name + "!"
        print(msg)
usernames= ['hannah', 'ty','margot']
greet_users(usernames)


# Allow a function to modify a list:
# the following examples sends a list of model to a function for printing.
# the original list is emptied and the second list is filled.
def print_models(unprinted, printed):
    while unprinted:
        current_model = unprinted.pop()
        print("printing " + current_model)
        printed.append(current_model)

# Store some unprinted designs, and print each of them.
unprinted = ['phone case', 'pendant', 'ring']
printed= []
print_models(unprinted, printed)

print("\nUnprinted:", unprinted)
print("printed:", printed)

# Prevent a function from identifying a list:
# the original list is unchanged after calling print_models.
original= ['phone case', 'pendant', 'ring']
printed2= []
print_models(original[:], printed2)
print("\noriginal:", original)
print("printed:", printed2)
```

```python
# Pass an arbitrary number of arguments:
# collect a arbitrary number of arguments (wont know how many arguments a function will need) into one parameter using operators.
# a parameter must come last in the function definition.
# the ** operator allows a function to collect arbitrary number of keyword arguments.

# Collect an arbitrary number of arguments:
def make_pizza(size, *toppings):
    print("\nMaking a " + size + "pizza.")
    print("toppings:")
    for topping in toppings:
        print("- " + topping)
# make three pizzas with different toppings -> pass many values
make_pizza('small', 'pepperoni')
make_pizza('large', 'bacon bits', 'pineapple')
make_pizza('medium', 'mushroom', 'peppers', 'onions','extra cheese')


# Collect an arbitrary number of keywords:
def build_profile(first, last, **user_info):
    profile = {'first': first, 'last': last}
    for key, value in user_info.items():
        profile[key] = value
    return profile
# create two users with different kinds of information -> phai truyen vao dung keyword moi chap nhan
user_0 = build_profile('albert','einstein', location = 'princeton')
user_1 = build_profile('marie', 'curie', location = 'paris', field= 'chemistry')
print(user_0)
print(user_1)

# kwargs prevent errors
def solve(**kwargs):
  a = kwargs.get('a', None)
  b = kwargs.get('b', None)
  c = kwargs.get('c', None)
  return a, b, c

solve(a=1, b=2, c=3)
solve(a=1, b=2)
```

```python
# Sort function
def sumDigit(n):
  res = 0
  while n != 0 :
    res += n % 10
    n //= 10
  return res

if __name__ == '__main__':
  testList = [3, 24, 111, 30, 12, 21, 99, 32]
  testList.sort(key = sumDigit)
  print(testList)

test_list = [-10, 230, 5, -4, -3, 10]
test_list.sort(key = lambda x : abs(x), reverse = True)
print(test_list)

test_list2 = [[1, 2], [3, 1], [3, 4], [0, 5], [1, 3]]
test_list2.sort(key = lambda x : (x[0], -x[1])) #first element increase else second element decrease
print(test_list2)

dic = [
{'name' : 'Tran Xuan Loc', 'job' : 'Dev', 'salary' : 500},
{'name' : 'Thieu Ngoc Tuan', 'job' : 'Dev', 'salary' : 1500},
{'name' : 'Phung Duc Kien', 'job' : 'BA', 'salary' : 500},
{'name' : 'Huynh Manh Tuong', 'job' : 'Tester', 'salary' : 2000}
]
dic.sort(key = lambda x : (x.get('salary'), x.get('name'))) #salary increase first else name increase
for x in dic:
  print(x)

# itemgetter (attrgetter)
from operator import itemgetter
friends = [("Huynh Manh Tuong", 30),
("Tran Nhat Phi", 28),
("Tuan Binh", 21),
("Thieu Tuan", 35)]

sort_friend = sorted(friends, key=itemgetter(0, 1)) #index of element in tuple
for x in friends:
  print(x)
```
