# List

```python
# Lists
# stores a series of items in a order.
# access through index within a loop.
bikes = ['trek', 'redline', 'giant', 1, 4, 2, 3]
first_bike = bikes[0]
last_bike = bikes[-1]

# Methods
numbers = [3, 8, 10, 2, 1, 7, 7, 8]
print(numbers.count(7))
numbers.sort()
numbers.reverse() #numbers.sort(reverse=True)
print(numbers)
even_number = 2
print(numbers.index(even_number)) # index of element in list

# built-in function
"""
all() Trả về True nếu mọi phần tử trong list đều là True
any() Trả về True nếu có ít nhất 1 phần tử trong list là True
max() Trả về phần tử lớn nhất trong list
min() Trả về phần tử nhỏ nhất trong list
sorted() Trả về list đã sắp xếp của list ban đầu
sum() Trả về tổng các phần tử trong list
"""
```

```python
# List constructor
s = "28tech"
b = list(s)
print(b, len(b), sep=' - ')
```

```python
# For each
print(first_bike, last_bike, sep='-')
for bike in bikes:
    print(bike, end=' ')
print()
```

```python
# Copy a list => lay gia tri va sao chep vao list moi neu dung "=" là 2 bien dung chung vung nho => doi a se anh huong b
copy_of_bikes = bikes[:]
print(copy_of_bikes)
# lap lai cac phan tu
copy_bikes = bikes * 2
print(copy_bikes)
# tao list 10 phan tu 0
zero = [0] * 10
print(zero)

# Catenation
bikes.append("cars") # them 1 phan tu
bikes.extend(b) # them nhieu phan tu
copy_of_bikes += b
print(bikes == copy_of_bikes)

# Search in list
if "motor" in bikes:
  print("YES")
else:
  print("NO")

# Create list
squares = []
for x in range (1,11):
    squares.append(x**2)
print(squares)
```

```python
# Modify list
squares.insert(2, 25)
squares[3] = 70
print(squares)

squares.pop() # last element but still use it when assigning into variable
squares.pop(0) # first element
squares.remove(25) # value
print(squares)

del squares[0] # phan tu cu the
print(squares)

squares.clear() # all list
print(squares)
```

```python
# List comprehensions
squares2 = [ x**2 for x in range (11)]
print(squares2)
check = [x % 2 == 0 for x in squares2]
print(check)

nums = [1, 2, 3, -5, 3, -4, 0]
num = [x for x in nums if x >= 0]
print(num)

n = 10
even_numbers = [x for x in range(0, n, 2)]
print(even_numbers)

z = range(10)
# listcomp
y = [x ** 2 for x in z]
print(y)
# map
c = list(map(lambda x : x ** 2, z))
print(c)

# listcomp
even = [x for x in z if x % 2 == 0]
print(even)
# map
fil = list(filter(lambda x : x % 2 == 0, z))
print(fil)
```

```python
# List slicing
# list[start, end, step]
finishers=['kai','abe', 'ada', 'gus', 'zoe']
first_three = finishers[ :3]
middle_three = finishers[1:4]
last_three = finishers[-3:]
reverse_list = finishers[::-1]
print(first_three)
print(middle_three)
print(last_three)
print(reverse_list)
upper_name = [x.upper() for x in finishers]
print(upper_name)
```

# Nested List - Mảng hai chiều

```python
# Input - output
n, m = 2, 3
a = [[0 for _ in range(m)] for _ in range(n)]
for i in range(n):
  a[i] = list(map(int, input().split()))
for i in range(n):
  for j in range(m):
    print(a[i][j], end = ' ')
  print()

# Index
print(a[0][1])
```
