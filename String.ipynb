# String

```python
# String
s = '28tech' #single quotes => s[0] = '7' cannot be change because it is immutable
t = "python java C++" #double quotes
#Multiline string
u = """
cuoc song
tuoi dep
"""
print(s, t)
print(u)

# Các tính năng tương tự như list, tuple, dictionary
# replace
str1 = t.replace('java', 'pascal')
print(str1)

# split
st1 = "28tech    python C++"
a = st1.split()
print(a)

st2 = "28tech,python,C++"
b = st2.split(",")
print(b)

# join
st3 = ["28tech", "python", "C++"]
t1 = ','.join(st3)
print(t1)

u1 = '##'.join(st3)
print(u1)

# substring
print("28tech" in s) #boolean
print(s.find("28tech")) #number

# string format
s1 = '%s is %d years old' % ('CR7', 37)
t11 = 'Pi = %.2f' % (3.14159)
u11 = '{0} is {1} years old !'.format('CR7', 37)
print(s1)
print(t11)
print(u11)
```
